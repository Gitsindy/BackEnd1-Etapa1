GRÁFICOS ESTADÍSTICOS

```python
sns.countplot(x='Municipio De Oferta Del Programa', hue='Id_Nucleo', data=df)
plt.xticks(rotation=90)
df = load_iris(as_frame=True)
df = df.frame
plt.title('Cantidad de graduados por Nucleo')
plt.xlabel('Municipio De Oferta Del Programa')
plt.ylabel('Cantidad de graduados')
plt.show()
plt.show()
```

```python
# Agrupar los datos para obtener la cantidad de graduados por sector y sexo
graduados_por_sector_sexo = df.groupby(['Sector IES', 'Sexo']).sum().reset_index()
# Gráfico de barras para visualizar los datos

plt.figure(figsize=(8, 6))
sns.barplot(data=graduados_por_sector_sexo, x='Sector IES', y='Graduados', hue='Sexo')
plt.title('Cantidad de Graduados en Universidades Públicas y Privadas por Sexo')
plt.xlabel('Tipo de Universidad')
plt.ylabel('Cantidad de Graduados')
plt.show()
```

```python
graduados_por_año = df.groupby(['Año', 'Sector IES']).sum().reset_index()

# Gráfico de líneas
plt.figure(figsize=(12, 6))
sns.lineplot(data=graduados_por_año, x='Año', y='Graduados', hue='Sector IES', marker='o')
plt.title('Evolución de Graduados por Año y Tipo de Universidad')
plt.xlabel('Año')
plt.ylabel('Cantidad de Graduados')
plt.show()
```

```python

# Agrupar los datos por año y sexo, y sumar los graduados
graduados_por_ano_sexo = df.groupby(['Año', 'Sexo'])['Graduados'].sum().reset_index()

# Crear el gráfico de barras
plt.figure(figsize=(12, 6))
barplot = sns.barplot(data=graduados_por_ano_sexo, x='Año', y='Graduados', hue='Sexo', palette='pastel')

# Añadir los valores encima de cada barra
for p in barplot.patches:
    barplot.annotate(
        format(p.get_height(), '.0f'),  # Formato del número (sin decimales)
        (p.get_x() + p.get_width() / 2., p.get_height()),  # Posición
        ha = 'center', va = 'center',  # Alineación horizontal y vertical
        xytext=(0, 8),  # Desplazamiento en píxeles
        textcoords='offset points'
    )

# Configuración del gráfico
plt.title('Total de Graduados por Sexo por Año (2015-2023)')
plt.xlabel('Año')
plt.ylabel('Cantidad de Graduados')
plt.legend(title='Sexo')
plt.show()
```

```python
# Agrupar los datos por sexo y sumar el total de graduados en el rango de años
graduados_por_sexo = df.groupby('Sexo')['Graduados'].sum().reset_index()

# Gráfico de barras para mostrar el total de graduados por sexo en el rango 2015-2023
plt.figure(figsize=(8, 6))
sns.barplot(data=graduados_por_sexo, x='Sexo', y='Graduados', palette='pastel')
for i, v in enumerate(graduados_por_sexo['Graduados']):
   plt.text(i, v + 1, str(v), ha='center')
plt.title('Total de Graduados por Sexo (2015-2023)')
plt.xlabel('Sexo')
plt.ylabel('Cantidad de Graduados')
plt.show()
```

```python
# núcleo de conocimiento
graduados_nbc = df.groupby('Núcleo Básico Del Conocimiento (NBC)').sum().reset_index().sort_values(by='Graduados', ascending=False)

# Visualizar los 10 núcleos más altos
plt.figure(figsize=(12, 8))
sns.barplot(data=graduados_nbc.head(10), x='Graduados', y='Núcleo Básico Del Conocimiento (NBC)')
plt.title('Top 10 Núcleos Básicos del Conocimiento por Cantidad de Graduados')
plt.xlabel('Cantidad de Graduados')
plt.ylabel('Núcleo Básico del Conocimiento')
plt.show()
```
